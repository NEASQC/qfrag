### 1 - Imports

```python
import qfrag as qf
import networkx as nx
import matplotlib.pyplot as plt
```

### 2 - Converting Between Formats

```python
mol_name = 'sdf/cannabidiol.sdf'
qf.molecule_converter(mol_name,'sdf','smi')
```

### 3 - Files to Graphs

```python
atoms, edges, bonds = qf.sdf2graph(mol_name)
print('Atoms:\n\n', atoms)
print('\nEdges:\n\n', edges)
print('\nBonds:\n\n', bonds)
```

### 4 - Graphs to Plots

```python
G = qf.graph2plot(atoms, edges, bonds)
```

```python
nx.draw_kamada_kawai(nx.minimum_spanning_tree(G))
```

```python
list(nx.chain_decomposition(G))
```
